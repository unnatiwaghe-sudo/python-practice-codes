```python
s1="SBJain college 456"

c=0
c1=0
c2=0
for i in s1:
  if i.isalnum():
    c+=1
  if i.isdigit():
      c1+=1
  if i.isalpha():
    c2+=1
print(c)
print(c1)
print(c2)
```

```python
s="welcome to python programming"
print(s.find("p"))#find first index  position of p ie: 11
print(s.index("p"))#also same
print(s.find("p",12))#find p from 12 index to end
print(s.index("p",12))#same like find
print(s.find("p",1,13))#find p from 1 to 12 index
print(s.index("p",1,13))#same like find
#difference in both
print(s.find("k"))  #give -1
#print(s.index("k")) #give error value error
#note find function only available with string
# not available in tupple,list,set but index function available
```

```python
s="Hello python Programming"
vowels = "aeiouAEIOU"
count = 0

for char in s:
    if char in vowels:
        count += 1

print("vowel: ",count)
```

```python
 #shows append,insert,extent function of list
my_list1=[1,"hindi",33,44,"edurekha"]
my_list1.append([555,22] )    #USE OF APPEND FUNCTION
print(my_list1)
print(len(my_list1))
my_list1.extend([444,666])    #USE OF EXTEND FUN
print(my_list1)
print(len(my_list1))
my_list1.insert(1,"sanjay")    #USE OF INSERT FUNCTION
print(my_list1)
print(len(my_list1))
print(my_list1 + [ "katre"])   #add content to given list but original list not effect
print(my_list1)
print(my_list1 *2)   #list is multiply by 2 shows list 2 times
print(my_list1)
my_list=[11,22,33,44]
print(my_list)
my_list.insert(-1,99)
print(my_list)
```

```python
#list delete function use ,remove,clear
my_list1=[1, 'sanjay', 'hindi', 33, 44, 'edurekha']
del my_list1[1]     #USE OF DELETE FUNCTION
print(my_list1)
my_list1.remove("hindi")  #REMOVE  ELEMENT FROM LIST IF NOT PRESENT GIVE ERROR
print(my_list1)
a= my_list1.pop(3)
print("pop element:",a,"list remaining:",my_list1)
my_list1.clear()
print(my_list1)
```
