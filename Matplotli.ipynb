```python
#matplotlib library practicval 11
import matplotlib.pyplot as plt
sales=[210,270,300,350,400,300,200,600]
year=[2011,2012,2013,2014,2015,2016,2017,2018]
plt.plot(year,sales)
plt.title("sales of john's company")
plt.xlabel("year")
plt.ylabel("sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
pen=[210,270,300,350,400,300,200,600]
copy=[300,320,360,400,200,120,300,350]
year=[2011,2012,2013,2014,2015,2016,2017,2018]
plt.plot(year,pen,color="yellow",label="pen",linestyle="-",linewidth=2,marker="o",markersize=5)
plt.plot(year,copy,color="green",label="copy",linestyle=":",linewidth=4,marker="o",markersize=5)
plt.title("sales of johnson company")
plt.xlabel("year")
plt.ylabel("sales")
plt.legend()  #shows coner  points x` is and y is
plt.show()
```

```python
#BAR GRAPH LEARN

import matplotlib.pyplot as plt
Name=["ssuman","rakesh","sanjay","vinod","raavi","yogesh","umesh"]
marks=[33,22,11,55,77,99,55]
plt.bar(Name,marks)
plt.title("Result")
plt.xlabel("Name")
plt.ylabel("marks")
plt.legend()
plt.show()
```

```python
#use of x limit
import matplotlib.pyplot as plt
Name=["ssuman","rakesh","sanjay","vinod","raavi","yogesh","umesh"]
marks=[33,22,11,55,77,99,55]
plt.bar(marks,Name,color="red",width=0.9) #same like as above bar also parameters wiidth by default 0.8
plt.title("Result")
plt.ylabel("Name")
plt.xlabel("marks")
plt.xlim(11,55)   #only shows upto daat of range between 11 to 55
#plt.xlim(50,100)   #only shows those data between 50 to 100
plt.legend()
plt.show()
```

```python
study_hours = [1, 2, 3, 4, 5]
marks = [20, 40, 60, 80, 100]

# Create the plot
plt.plot(study_hours, marks, label="Marks and hours",
         linestyle='-', color='blue', marker='o')

# Add title and labels
plt.title("Study Hours vs Marks")
plt.xlabel("No. of hours")
plt.ylabel("Marks")

# Add legend
plt.legend()

# Display the plot
plt.grid(True) # adds a grid to the graph.its shows grid
plt.show()
```
