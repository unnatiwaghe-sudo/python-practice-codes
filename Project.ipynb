```python
class Node:
    def __init__(self, pid, size, status):
        self.pid = pid
        self.size = size
        self.status = status
        self.next = None


class MemoryManager:
    def __init__(self, total_memory):
        self.head = Node(None, total_memory, "Free")

    def allocate(self, pid, size):
        temp = self.head

        while temp is not None:
            if temp.status == "Free" and temp.size >= size:
                remaining = temp.size - size

                temp.pid = pid
                temp.size = size
                temp.status = "Allocated"

                if remaining > 0:
                    new_node = Node(None, remaining, "Free")
                    new_node.next = temp.next
                    temp.next = new_node

                print("Memory allocated to Process", pid)
                return

            temp = temp.next

        print("Memory allocation failed")

    def deallocate(self, pid):
        temp = self.head

        while temp is not None:
            if temp.pid == pid:
                temp.pid = None
                temp.status = "Free"
                print("Memory deallocated from Process", pid)
                return

            temp = temp.next

        print("Process not found")

    def display(self):
        temp = self.head

        print("\nMemory Status:")
        print("-----------------------------")

        while temp is not None:
            print("PID:", temp.pid, "| Size:", temp.size, "| Status:", temp.status)
            temp = temp.next

        print("-----------------------------")


total_memory = int(input("Enter Total Memory Size: "))
mm = MemoryManager(total_memory)

while True:
    print("\nMemory Allocation Simulator")
    print("1. Allocate Memory")
    print("2. Deallocate Memory")
    print("3. Display Memory")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        pid = int(input("Enter Process ID: "))
        size = int(input("Enter Memory Size: "))
        mm.allocate(pid, size)

    elif choice == 2:
        pid = int(input("Enter Process ID to deallocate: "))
        mm.deallocate(pid)

    elif choice == 3:
        mm.display()

    elif choice == 4:
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice")
```
