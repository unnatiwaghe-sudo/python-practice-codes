```python
#dictionary of lists example
#NOW WE CREATE DATAFRAME USING DICTIONARY
import pandas as pd
a1=["11A","12B","11C","12D"]
a2=["sanju","umesh","raju","viju"]
a3=[90,80,70,66]
dict={"Class":a1,"Name":a2,"Marks":a3}   #keys become column in dictionary
s=pd.DataFrame(dict)
print(s)
```
